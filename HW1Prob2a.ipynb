```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def sigmoid(z):
    # Computes the sigmoid activation function.

    return 1 / (1 + np.exp(-z))

def hard_limit(z):
       
    return (z >= 0).astype(int)

def radial_basis_function(x1, x2, center=(0, 0), sigma=1):
    # Computes a Radial Basis Function (RBF) as an activation function.
    
    dist_sq = (x1 - center[0])**2 + (x2 - center[1])**2
    return np.exp(-dist_sq / (2 * sigma**2))

def perceptron_output(x1, x2, activation_func, rbf_mode=False):
    # Calculates the output of the perceptron based on the given activation function.
    
    if rbf_mode:
        # For RBF, the inputs x1 and x2 are passed directly to the function
        return activation_func(x1, x2)
    else:
        # For sigmoid and hard limit, the weighted sum is the input
        z = -4.79 * x1 + 5.90 * x2 - 0.93
        return activation_func(z)

def plot_surface(activation_name, num_samples, activation_func, rbf_mode=False):
   
    # Define the domain for x1 and x2
    domain_min = -2
    domain_max = 2

    # Generate random sample points for x1 and x2 within the domain
    x1_rand = np.random.uniform(domain_min, domain_max, num_samples)
    x2_rand = np.random.uniform(domain_min, domain_max, num_samples)

    # Calculate the corresponding output y for each sample point
    y = perceptron_output(x1_rand, x2_rand, activation_func, rbf_mode)

    # Create a 3D plot
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    ax.plot_trisurf(x1_rand, x2_rand, y, cmap='viridis')

    # Set labels and title for the plot
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_zlabel('$y$')
    ax.set_title(f'Perceptron Output Surface with {activation_name} Activation ({num_samples} Samples)')

    plt.show()

if __name__ == "__main__":
    sample_sizes = [100, 5000, 10000]

    # Problem (a): Sigmoid Activation
    for num_samples in sample_sizes:
        plot_surface("Sigmoid", num_samples, sigmoid, rbf_mode=False)
```

```python
    # Problem (b): Hard Limit Activation
    for num_samples in sample_sizes:
        plot_surface("Hard Limit", num_samples, hard_limit, rbf_mode=False)
```

```python
    # Problem (c): Radial Basis Function (RBF) Activation
    for num_samples in sample_sizes:
        plot_surface("Radial Basis Function", num_samples, radial_basis_function, rbf_mode=True)
```
